#Set

```python
#Cretion of set
myset = {"maths", "phy", "chem"}
print("Initial set:", myset)
```

```python
#dupliates are not allowed in sets
myset.add("maths")
print("Set after trying to add dupliate element maths:", myset)
```

```python
#The values True and 1 are considered the same val;ue in sets, and are treated as duplictes
myset = {"maths", "phy", "chem", True, 1 ,2}
print(myset)
```

```python
#A et ith string integers are boolean values
set1 = {"xyz", 78, True, 38, "female", 11.3}
set1
```

```python
#To check membership
print("maths" in myset)
print("chem" in myset)
```

```python
#once a set is created, we cannot change its elements, but we can add elements
myset.add("eng")
myset
```

```python
#remove item
myset = {"maths", "phy", "chem"}
myset.remove("phy")
myset
```

```python
#Set Operations
C = {2, 3, 4, 6, 8, 9}
D = {1, 2, 3, 5, 7, 9}

print("Union:", C|D)
print("Intersection:", C & D)
print("Difference:", C - D)
print("Symmetric Difference:", C^D)
```

#Python Dictionaries

```python
#Creation of dictionary
mydict = {
    "name": "Jahnvi",
    "fruit" : "apple",
    "meal" : "dinner"
}
print(mydict)
```

```python
#Dictionary takes latest value given for key. It will not allow duplicates
mydict = {
    "name": "Jahnvi",
    "fruit" : "apple",
    "meal" : "dinner",
    "meal": "lunch"
}
print(mydict)
```

```python
#Length of dictionary
mydict = {
    "name": "Jahnvi",
    "fruit" : "apple",
    "meal" : "dinner",
    "meal": "lunch"
}
print("Length of this dict", len(mydict))
```

```python
#Class of dictionary
mydict = {
    "name": "Jahnvi",
    "fruit" : "apple",
    "meal" : "dinner",
    "meal": "lunch"
}
print(type(mydict))
```

```python
#Changing/Adding key, values in dictionary
mydict = {
    "name": "Jahnvi",
    "fruit" : "apple",
    "meal" : "dinner"
}
y = mydict.values()
mydict["colour"] = "blue"
print(y)
```
